### Mathematical review of Kinematics

````python
%matplotlib tk
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

```
Joint | a (m)     | d (m)    | Description
------+-----------+----------+------------------------------------------
  1   |  0        | 0.15185  | Base height (from floor to joint 2 axis)
  2   | -0.24355  | 0        | Upper arm length
  3   | -0.21325  | 0        | Forearm length
  4   |  0        | 0.13105  | Wrist 1 offset
  5   |  0        | 0.08535  | Wrist 2 offset
  6   |  0        | 0.0921   | Tool flange offset
```

````python
def dh_matrix(theta, d, a, alpha):
    ct, st = np.cos(theta), np.sin(theta)
    ca, sa = np.cos(alpha), np.sin(alpha)
    return np.array([
        [ct, -st*ca,  st*sa, a*ct],
        [st,  ct*ca, -ct*sa, a*st],
        [0,   sa,     ca,    d   ],
        [0,   0,      0,     1   ]
    ])
````

````python
dh_params = [
    [np.radians(0), 0.15185, 0, np.radians(90)],
    [np.radians(0), 0, -0.24355, np.radians(90)],
    [np.radians(0), -0.21325, 0, np.radians(90)],
    [np.radians(0), 0, 0.13105, np.radians(90)],
    [np.radians(0), 0.08535,0, np.radians(0)],
    [np.radians(0), 0.0921, 0, np.radians(0)]
]
````

````python
dh_params[5][1] += 0.1
````

````python
# Compute forward kinematics
positions = [np.array([0, 0, 0])]  # base origin
T_cumulative = np.eye(4)

for params in dh_params:
    T_cumulative = T_cumulative @ dh_matrix(*params)
    pos = T_cumulative[:3, 3]
    positions.append(pos)

positions = np.array(positions)
````

````python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(positions[:, 0], positions[:, 1], positions[:, 2], 'o-', linewidth=3, markersize=8, color='royalblue')
ax.scatter(*positions[0], color='green', s=100, label='Base')
ax.scatter(*positions[-1], color='red', s=100, label='End-effector')

ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')
ax.set_title('UR3e - Joint 1 only')
ax.legend()

# Set equal aspect ratio for all axes
max_range = 0.3
ax.set_xlim([-max_range, max_range])
ax.set_ylim([-max_range, max_range])
ax.set_zlim([0, max_range * 2])

plt.tight_layout()
plt.show()
````
