---- Create trajectory ----

````python
# from src.transformation import tcp_trans
traces = [data["traces"][i] for i in [1, 4]]  # keep only traces 1 and 4
draw_motions = []
move_motions = []
above = [0,0,-0.02,0,0,0]
for trace in traces:
    normal = trace["face"]
    t_world = trace["path"]
    t_world = t_world[:3]  # limit to first 3 points per face for faster iteration
    motion = []
    
    normal_stl = obj_to_stl(np.array(normal))

    # Drawing motion
    for i, p in enumerate(t_world):
        pt_stl = obj_to_stl(np.array(p))
        p_w = (*pt_stl, *normal_stl)
        x,y,z,rx,ry,rz = obj2robot(p_w)
        
        # Add entry point
        if i == 0:
            x_t,y_t,z_t,_,_,_ = tcp_trans([x,y,z,rx,ry,rz], above)
            m = TCP6D.createFromMetersRadians(x_t,y_t,z_t, rx,ry,rz)
            motion.append(m)

        # Add drawing point
        m = TCP6D.createFromMetersRadians(x,y,z, rx,ry,rz)
        motion.append(m)

        # Add exit point
        if i == len(t_world)-1:
            x_t,y_t,z_t,_,_,_ = tcp_trans([x,y,z,rx,ry,rz], above)
            m = TCP6D.createFromMetersRadians(x_t,y_t,z_t, rx,ry,rz)
            motion.append(m)
    
    if motion:
        draw_motions.append(motion)
        move_motions.append((motion[0], motion[-1]))
````

````python
from src.safety import CollisionChecker
from scipy.spatial.transform import Rotation as Rot
import pybullet as pb

# Extract position and orientation from the obj→robot transformation
T = obj2robot.T
pos = T[:3, 3].tolist()
````

````python
print(pos)
````

````python

# R contains rotation + mm→m scaling; remove scale to get pure rotation
R = T[:3, :3]
scale = np.linalg.norm(R[:, 0])        # ~0.001 (uniform scale)
R_pure = R / scale
quat = Rot.from_matrix(R_pure).as_quat().tolist()  # [x, y, z, w] — PyBullet format

# gui=True opens a PyBullet 3D viewer window
# STL is in millimeters — scale to meters
checker = CollisionChecker(
    obstacle_stls=[{
        'path': 'my_simulation/3d_objects/trapez.stl',
        'scale': [0.001, 0.001, 0.001],
        'position': pos,
        'orientation': quat,
    }],
    gui=True,
)
print(f"PyBullet GUI running (cid={checker.cid})")
print(f"Robot body id: {checker.robot_id}")
print(f"Obstacle ids:  {checker.obstacle_ids}")

# Print obstacle position and bounding box
for oid in checker.obstacle_ids:
    pos_ob, orn = pb.getBasePositionAndOrientation(oid, physicsClientId=checker.cid)
    aabb_min, aabb_max = pb.getAABB(oid, physicsClientId=checker.cid)
    print(f"\nObstacle {oid}:")
    print(f"  Position:    {pos_ob}")
    print(f"  Orientation: {orn}")
    print(f"  AABB min:    {[f'{v:.4f}' for v in aabb_min]}")
    print(f"  AABB max:    {[f'{v:.4f}' for v in aabb_max]}")
    print(f"  Size (m):    ({aabb_max[0]-aabb_min[0]:.4f}, {aabb_max[1]-aabb_min[1]:.4f}, {aabb_max[2]-aabb_min[2]:.4f})")
````

### Test validate_tcp on specific poses

````python
# Test 1: Home position should be valid
ok, q, reason = checker.validate_tcp(robot_sim, tcp_home)
print(f"Home TCP valid: {ok}  (reason: {reason or 'none'})")

# Test 2: A point with Y > 0 should be rejected (workspace bounds)
bad_tcp = TCP6D.createFromMetersRadians(0.0, 0.1, 0.2, 0, np.pi, 0)
ok, q, reason = checker.validate_tcp(robot_sim, bad_tcp)
print(f"Y>0 TCP valid:  {ok}  (reason: {reason})")

# Test 3: A point with Z > 0.5 should be rejected
bad_tcp2 = TCP6D.createFromMetersRadians(0.0, -0.2, 0.6, 0, np.pi, 0)
ok, q, reason = checker.validate_tcp(robot_sim, bad_tcp2)
print(f"Z>0.5 TCP valid: {ok}  (reason: {reason})")
````

````python
import colorsys
import pybullet

# Generate distinct colors for each trace
def trace_colors(n):
    return [colorsys.hsv_to_rgb(i / max(n, 1), 0.9, 1.0) for i in range(n)]

colors = trace_colors(len(draw_motions))

# --- Phase 1: precompute all IK solutions ---
print("Solving IK for all waypoints...")
all_solutions = []  # list of lists of (wp_idx, q_list, tcp_list) or None
for trace_idx, trace in enumerate(draw_motions):
    trace_sols = []
    prev_q = None
    for wp_idx, tcp in enumerate(trace):
        q = robot_sim.get_inverse_kin(tcp, qnear=prev_q)
        if q is None:
            coords = tcp.toList()
            print(f"  [SKIP] trace {trace_idx} wp {wp_idx}: IK failed for ({coords[0]:.4f}, {coords[1]:.4f}, {coords[2]:.4f})")
            trace_sols.append(None)
        else:
            trace_sols.append(q.toList())
            prev_q = q
    all_solutions.append(trace_sols)
print(f"IK done. {sum(s is not None for t in all_solutions for s in t)}/{sum(len(t) for t in all_solutions)} solved.")

# --- Phase 2: instant teleport + draw debug lines ---
for trace_idx, (trace, sols) in enumerate(zip(draw_motions, all_solutions)):
    rgb = colors[trace_idx]
    prev_pos = None
    for wp_idx, (tcp, q_list) in enumerate(zip(trace, sols)):
        if q_list is None:
            prev_pos = None
            continue
        checker.set_joint_angles(q_list)

        is_drawing = 0 < wp_idx < len(trace) - 1
        cur_pos = tcp.toList()[:3]
        if is_drawing and prev_pos is not None:
            pybullet.addUserDebugLine(prev_pos, cur_pos, rgb, lineWidth=3,
                                      physicsClientId=checker.cid)
        prev_pos = cur_pos if is_drawing else None
    print(f"Trace {trace_idx + 1}/{len(draw_motions)} drawn")
````

### Test free-space path planning

````python
from src.pathfinding import find_path
import time

# Plan a path between first and second trace hover points
if len(draw_motions) >= 2:
    start_tcp = draw_motions[0][-1]   # exit of trace 0
    end_tcp   = draw_motions[1][0]    # entry of trace 1
    
    print(f"Planning path from ({start_tcp.x:.3f}, {start_tcp.y:.3f}, {start_tcp.z:.3f})")
    print(f"               to   ({end_tcp.x:.3f}, {end_tcp.y:.3f}, {end_tcp.z:.3f})")
    
    waypoints = find_path(robot_sim, checker, start_tcp, end_tcp)
    print(f"Path found: {len(waypoints)} waypoints")
    
    # Animate in PyBullet
    for i, wp in enumerate(waypoints):
        q = robot_sim.get_inverse_kin(wp)
        if q is not None:
            checker.set_joint_angles(q.toList())
            time.sleep(0.5)
            print(f"  wp {i}: ({wp.x:.3f}, {wp.y:.3f}, {wp.z:.3f})")
    print("Done.")
````
